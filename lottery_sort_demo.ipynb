# 員工樂透抽號排序模擬

目的
透過逐步抽出樂透號碼，
決定員工領獎順序。

特色
不是一次抽籤決定，
而是透過多輪抽號逐步產生排序。


## 抽號規則

樂透號碼範圍
1 ~ 30

19位員工簽注
4 個不重複號碼

主持人抽號

第一次抽
5 個號碼

之後每輪
加抽 1 個號碼

當有序區員工 >= 9
停止抽號

## 排序規則

1 對中球數多者優先
2 球數相同 → 比對中號碼大小(現場抽比大或比小)
3 比對號碼相同 → 各自再抽一球(現場抽比大或比小)



```
##初始化設定（Python）
初始化員工
初始化樂透號碼
初始化區域
```



````python
import random
import time
from IPython.display import clear_output

# =========================
# 初始化設定
# =========================

employees = ["宛","馬","周","琇","袋","孟","懿","定","厚","企","姿","書","柏","慶","欣","佩","曼","德","勇"]

lottery_pool = list(range(1,31))

unordered = employees.copy()
ordered = []

employee_numbers = {}

# =========================
# 開場畫面
# =========================

print("🎰 員工樂透排序抽獎系統")
print("----------------------")
time.sleep(2)

print("\n初始化狀態")
print("普獎區員工：", unordered)
print("大獎區員工：", ordered)

time.sleep(3)

# =========================
# 員工簽注號碼（動畫）
# =========================

print("\n🎟️ 員工開始簽注樂透號碼\n")
time.sleep(2)

for emp in employees:

    print(f"✍️ {emp} 簽注中...")
    time.sleep(0.5)

    numbers = random.sample(lottery_pool,4)
    numbers.sort()

    employee_numbers[emp] = numbers

    print(f"🎯 {emp} 的號碼：{numbers}\n")

    time.sleep(1)

print("所有員工簽注完成 ✅")
````

````python
import random
import time

print("\n🎤 主持人準備抽出樂透號碼")
time.sleep(2)

print("\n🎲 Round 1 開始")
time.sleep(2)

drawn_numbers = []

print("\n主持人正在搖獎球機...")
time.sleep(2)

# 第一次抽5顆球
for i in range(5):

    print("\n球機轉動中...")
    time.sleep(2)

    ball = random.choice([n for n in lottery_pool if n not in drawn_numbers])

    print(f"🎱 抽出號碼：{ball}")

    drawn_numbers.append(ball)

    time.sleep(2)

print("\n======================")
print("📢 本輪抽出號碼：")
print(sorted(drawn_numbers))
print("======================")
````

````python
import time

print("\n🔎 開始檢查誰對中號碼...\n")
time.sleep(2)

for emp in employees:

    emp_nums = employee_numbers[emp]

    matched = list(set(emp_nums) & set(drawn_numbers))

    time.sleep(1)

    if len(matched) > 0:

        print(f"🎯 {emp} 對中號碼 {sorted(matched)}")

        if emp not in ordered:

            ordered.append(emp)

            if emp in unordered:
                unordered.remove(emp)

            print(f"🏅 {emp} 進入大獎區！\n")

    else:

        print(f"{emp} 沒有對中號碼")

    time.sleep(0.5)

print("\n======================")
print("📋 當前區域狀態")
print("----------------------")

print("普獎區員工：")
print(unordered)

print("\n大獎區員工：")
print(ordered)

print("======================")
````

##加碼抽號循環

````python
import random
import time

round_num = 2

while len(ordered) < 9:

    print(f"\n🎲 Round {round_num}")
    print("主持人準備加抽號碼...")
    time.sleep(2)

    print("球機轉動中...")
    time.sleep(2)

    # 抽出一顆新的球
    new_ball = random.choice([n for n in lottery_pool if n not in drawn_numbers])

    drawn_numbers.append(new_ball)

    print(f"🎱 抽出號碼：{new_ball}")
    time.sleep(2)

    print("\n🔎 檢查誰對中號碼...\n")
    time.sleep(2)

    # 檢查每位員工
    for emp in employees:

        if emp in ordered:
            continue

        emp_nums = employee_numbers[emp]

        matched = list(set(emp_nums) & set(drawn_numbers))

        if len(matched) > 0:

            print(f"🎯 {emp} 對中 {sorted(matched)}")

            ordered.append(emp)
            unordered.remove(emp)

            print(f"🏅 {emp} 進入大獎區！\n")

            time.sleep(0.5)

        else:
            print(f"{emp} 仍未對中")
            time.sleep(0.5)

    print("\n📋 當前區域狀態")
    print("----------------------")

    print("普獎區員工：")
    print(unordered)

    print("\n大獎區員工：")
    print(ordered)

    print("----------------------")

    round_num += 1

    time.sleep(3)

print("\n🎉 大獎區已達 9 人，停止抽號！")
print("最終抽出號碼：", sorted(drawn_numbers))
````

以下是同球數群組比序列的版本，我發現1個BUG，每當顯群組比序列大小時： 若本輪規則比小時，排名都正確無誤 但本輪規則比大時 比較序列排序，每個人自己序列內的數字是正確排序的 FFF[29,18] EEE[30,2] 但人的排序卻錯了 應該要是 EEE[30,2] 確認第1名 FFF[29,18]確認第2名

````python
import random
import time
from collections import defaultdict

print("\n🏆 開始最終排序\n")
time.sleep(2)

# =========================
# Step1 計算命中資訊
# =========================

results = []

for emp in ordered:

    emp_nums = employee_numbers[emp]
    matched = sorted(list(set(emp_nums) & set(drawn_numbers)))

    results.append({
        "name": emp,
        "hit": len(matched),
        "numbers": matched[:]
    })

# =========================
# Step2 分群
# =========================

groups = defaultdict(list)

for r in results:
    groups[r["hit"]].append(r)

sorted_hits = sorted(groups.keys(), reverse=True)

rank_cursor = 1
group_rank_range = {}

print("📊 分群結果\n")

for h in sorted_hits:

    group = groups[h]
    start = rank_cursor
    end = rank_cursor + len(group) - 1

    group_rank_range[h] = (start, end)

    print(f"{h}球群組"+"{")

    for p in group:
        nums=",".join(map(str,p["numbers"]))
        print(f"{p['name']}[{nums}]")

    print("}")

    ranks=list(range(start,end+1))
    rank_str="、".join(map(str,ranks))

    print(f"~以上暫列第{rank_str}名\n")

    rank_cursor += len(group)

time.sleep(2)

# =========================
# Step3 由後往前排序(同球數比序列)
# =========================

final_ranking=[]

for hit in sorted(sorted_hits):

    group=groups[hit]
    start,end=group_rank_range[hit]

    print("\n-----------------------")
    print(f"{hit}球群組排序開始\n")

    if len(group)==1:

        p=group[0]
        nums=",".join(map(str,p["numbers"]))

        print(f"{p['name']}[{nums}] ~確定第{start}名")

        final_ranking=group+final_ranking
        continue

    # 抽籤決定比大或比小
    mode=random.choice(["big","small"])

    if mode=="big":
        print("🎲 本輪規則：比大\n")
    else:
        print("🎲 本輪規則：比小\n")

    # 建立比較序列
    for p in group:

        if mode=="big":
            p["compare"]=sorted(p["numbers"],reverse=True)
        else:
            p["compare"]=sorted(p["numbers"])

    # 排序
    group_sorted=sorted(group,key=lambda x:x["compare"])

    # 顯示排序用序列
    print("比較序列：")

    for p in group_sorted:

        seq=",".join(map(str,p["compare"]))
        print(f"{p['name']} → [{seq}]")

    print()

    i=0
    current_rank=start
    resolved=[]

    while i<len(group_sorted):

        tied=[group_sorted[i]]
        j=i+1

        while j<len(group_sorted) and group_sorted[j]["compare"]==group_sorted[i]["compare"]:
            tied.append(group_sorted[j])
            j+=1

        # 沒平手
        if len(tied)==1:

            p=tied[0]
            nums=",".join(map(str,p["numbers"]))

            print(f"{p['name']}[{nums}] ~確定第{current_rank}名")

            resolved.append(p)
            current_rank+=1

        else:

            rank_list=list(range(current_rank,current_rank+len(tied)))
            rank_str="、".join(map(str,rank_list))

            for p in tied:

                nums=",".join(map(str,p["numbers"]))
                print(f"{p['name']}[{nums}] ~暫列第{rank_str}名")

            print("\n⚠ 號碼完全相同")
            print("各自再抽1球\n")

            remaining=list(set(range(1,31))-set(drawn_numbers))

            for p in tied:

                ball=random.choice(remaining)
                p["numbers"].append(ball)
                p["numbers"]=sorted(p["numbers"])

                print(f"{p['name']} 抽到 🎱 {ball}")

                time.sleep(1)

            print()

            # 再抽籤
            mode2=random.choice(["big","small"])

            if mode2=="big":
                print("🎲 抽球比較規則：比大\n")
                for p in tied:
                    p["compare"]=sorted(p["numbers"],reverse=True)
            else:
                print("🎲 抽球比較規則：比小\n")
                for p in tied:
                    p["compare"]=sorted(p["numbers"])

            tied=sorted(tied,key=lambda x:x["compare"])

            print("比較序列：")

            for p in tied:
                seq=",".join(map(str,p["compare"]))
                print(f"{p['name']} → [{seq}]")

            print()

            for p in tied:

                nums=",".join(map(str,p["numbers"]))

                print(f"{p['name']}[{nums}] ~確定第{current_rank}名")

                resolved.append(p)
                current_rank+=1

        i=j

    final_ranking=resolved+final_ranking

time.sleep(2)

# =========================
# Step4 最終結果
# =========================

print("\n==========================")
print("🏆 最終排名")
print("==========================\n")

rank=1

for p in final_ranking:

    nums=",".join(map(str,p["numbers"]))

    print(f"{rank}️⃣ {p['name']}[{nums}]")

    rank+=1
    time.sleep(1)
````

````python
import random
import time
from collections import defaultdict

print("\n🏆 開始最終排序\n")
time.sleep(2)

# ======================
# Step1 計算命中
# ======================

results = []

for emp in ordered:

    emp_nums = employee_numbers[emp]
    matched = sorted(list(set(emp_nums) & set(drawn_numbers)))

    results.append({
        "name": emp,
        "hit": len(matched),
        "numbers": matched[:],
        "extra": []
    })

# ======================
# Step2 分群
# ======================

groups = defaultdict(list)

for r in results:
    groups[r["hit"]].append(r)

sorted_hits = sorted(groups.keys(), reverse=True)

rank_cursor = 1
group_rank_range = {}

print("分群結果\n")

for h in sorted_hits:

    group = groups[h]
    start = rank_cursor
    end = rank_cursor + len(group) - 1

    group_rank_range[h] = (start, end)

    print(f"{h}球群組"+"{")

    for p in group:
        nums = ",".join(map(str,p["numbers"]))
        print(f"{p['name']}[{nums}]")

    print("}")

    ranks = list(range(start,end+1))
    ranks_str = "、".join(map(str,ranks))

    print(f"~以上暫列第{ranks_str}名\n")

    rank_cursor += len(group)

time.sleep(2)

# ======================
# Step3 由後往前排序(此版分群只比最大1球或最小1球的號碼)
# ======================

final_ranking = []

for hit in sorted(sorted_hits):

    group = groups[hit]
    start,end = group_rank_range[hit]

    print("\n------------------")
    print(f"{hit}球群組排序開始")
    time.sleep(2)

    if len(group) == 1:

        p = group[0]
        nums=",".join(map(str,p["numbers"]))

        print(f"{p['name']}[{nums}] ~確定第{start}名")

        final_ranking = group + final_ranking
        continue

    mode=random.choice(["big","small"])

    if mode=="big":
        print("本輪規則：比大\n")
    else:
        print("本輪規則：比小\n")

    for p in group:

        if mode=="big":
            p["compare"]=max(p["numbers"])
        else:
            p["compare"]=min(p["numbers"])

    group_sorted=sorted(
        group,
        key=lambda x:x["compare"],
        reverse=(mode=="big")
    )

    # 找平手
    i=0
    resolved=[]
    current_rank=start

    while i<len(group_sorted):

        tied=[group_sorted[i]]
        j=i+1

        while j<len(group_sorted) and group_sorted[j]["compare"]==group_sorted[i]["compare"]:
            tied.append(group_sorted[j])
            j+=1

        if len(tied)==1:

            p=tied[0]
            nums=",".join(map(str,p["numbers"]))

            print(f"{p['name']}[{nums}] ~確定第{current_rank}名")

            resolved.append(p)
            current_rank+=1

        else:

            rank_list=list(range(current_rank,current_rank+len(tied)))
            rank_str="、".join(map(str,rank_list))

            for p in tied:
                nums=",".join(map(str,p["numbers"]))
                print(f"{p['name']}[{nums}] ~暫列第{rank_str}名")

            print("\n各自再抽1球\n")

            remaining=list(set(range(1,31))-set(drawn_numbers))

            for p in tied:

                ball=random.choice(remaining)
                p["extra"].append(ball)

                p["numbers"]=sorted(p["numbers"]+[ball])

                nums=",".join(map(str,p["numbers"]))
                print(f"{p['name']} 抽到 {ball}")

                time.sleep(1)

            mode2=random.choice(["big","small"])

            if mode2=="big":
                print("\n本輪規則：比大\n")
                tied=sorted(tied,key=lambda x:max(x["numbers"]),reverse=True)
            else:
                print("\n本輪規則：比小\n")
                tied=sorted(tied,key=lambda x:min(x["numbers"]))

            for p in tied:

                nums=",".join(map(str,p["numbers"]))

                print(f"{p['name']}[{nums}] ~確定第{current_rank}名")

                resolved.append(p)
                current_rank+=1

        i=j

    final_ranking=resolved+final_ranking

time.sleep(2)

# ======================
# 最終結果
# ======================

print("\n======================")
print("🏆 最終排名")
print("======================\n")

rank=1

for p in final_ranking:

    nums=",".join(map(str,p["numbers"]))

    print(f"{rank}️⃣ {p['name']}[{nums}]")

    rank+=1
````
